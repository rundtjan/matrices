```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read data from stored numpy-datafile
with open('result.npy', 'rb') as f:
    a = np.load(f)
    b = np.load(f)
```

```python
#Plot both CPU- and memory-utilization
plt.scatter(np.arange(stop=len(a[:4000])), a[:4000], s=0.5, label="CPU")
plt.scatter(np.arange(stop=len(b[:4000])), b[:4000], s=0.5, label="Memory")
plt.xlabel("Time (0.01s)")
plt.ylabel("Percentage")
plt.legend()
```
